#Data Types and Structures Questions

1. What are data structures, and why are they important?

Data structures are ways of organizing and storing data so it can be accessed and modified efficiently.
They are important because they improve performance, reduce complexity, and make programs easier to manage and understand.

2. Explain the difference between mutable and immutable data types with examples.

Mutable data types can be changed after creation (e.g., list, dict, set).

Immutable data types cannot be changed after creation (e.g., int, float, string, tuple).

Example:

lst = [1, 2]
lst.append(3)   # Allowed

s = "hi"
s[0] = "H"      # Error (immutable)

3. What are the main differences between lists and tuples in Python?

Lists are mutable, tuples are immutable

Lists use [], tuples use ()

Lists are slower; tuples are faster and safer

Tuples can be used as dictionary keys, lists cannot

4. Describe how dictionaries store data.

Dictionaries store data as key–value pairs using a hash table, allowing fast access to values using unique keys.

5. Why might you use a set instead of a list in Python?

Sets automatically remove duplicates

Faster membership testing (in)

Useful when uniqueness matters

6. What is a string in Python, and how is it different from a list?

A string is an immutable sequence of characters, while a list is a mutable sequence of elements.
Strings cannot be modified; lists can.

7. How do tuples ensure data integrity in Python?

Because tuples are immutable, their values cannot be changed accidentally, making them safe for fixed data.

8. What is a hash table, and how does it relate to dictionaries in Python?

A hash table stores data by converting keys into indexes using a hash function.
Python dictionaries use hash tables to provide fast lookups.

9. Can lists contain different data types in Python?

Yes. Lists can store multiple data types in a single list.

Example:

data = [1, "hello", 3.5, True]

10. Explain why strings are immutable in Python.

Strings are immutable to improve performance, memory efficiency, and security, especially when reused.

11. What advantages do dictionaries offer over lists for certain tasks?

Faster lookup using keys

Better organization of related data

Ideal for mapping relationships (e.g., name → age)

12. Describe a scenario where using a tuple would be preferable over a list.

When storing fixed data like coordinates or days of the week where values should not change.

13. How do sets handle duplicate values in Python?

Sets automatically remove duplicates and only store unique elements.

14. How does the in keyword work differently for lists and dictionaries?

In lists: checks for values

In dictionaries: checks for keys

15. Can you modify the elements of a tuple? Explain why or why not.

No. Tuples are immutable, so their elements cannot be changed after creation.

16. What is a nested dictionary, and give an example of its use case?

A dictionary inside another dictionary.
Used for structured data like student records.

Example:

students = {
  "Alice": {"age": 20, "grade": "A"}
}

17. Describe the time complexity of accessing elements in a dictionary.

Average time complexity is O(1) due to hashing.

18. In what situations are lists preferred over dictionaries?

When order matters

When accessing data by index

When storing simple sequences

19. Why are dictionaries considered unordered, and how does that affect data retrieval?

Dictionaries are conceptually unordered because data is stored by hash keys, not position.
Retrieval is done using keys, not indexes.

20. Explain the difference between a list and a dictionary in terms of data retrieval.

Lists retrieve data using index positions

Dictionaries retrieve data using keys, making lookups faster and more meaningful
